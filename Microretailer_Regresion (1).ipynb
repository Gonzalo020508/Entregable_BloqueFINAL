```python
#analizamos caso 1 de correlacin entre variable de nmero de clientes en la tienda respecto al nmero de estantes en la tienda, el nmero de refrigeradores y el nmero de empleados permanentes
Vars_Indep= dflimpio[['268_number_fridges','104_how_many_shelves_does_the_micro_retailer_have', '2_current_permanent_employees']]
Var_Dep= dflimpio['97_number_of_customers_in_store']
```

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R determinado
model.score(Vars_Indep,Var_Dep)
```

```python
#Declaramos las variables dependientes e independientes para la regresin logstica
Vars_Indep= dflimpio[['268_number_fridges','104_how_many_shelves_does_the_micro_retailer_have', '2_current_permanent_employees']]
Var_Dep= dflimpio['102_does_the_micro_retailer_sells_fresh_products_']
```

```python
#Redefinimos las variables
X= Vars_Indep
y= Var_Dep
```

```python
#Dividimos el conjunto de datos en la parte de entrenamiento y prueba:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =None)
```

```python
#Se escalan todos los datos
escalar= StandardScaler()
```

```python
#Para realizar el escalamiento de las variables "X" tanto de entrenamiento como de prueba, utilizaremos fit_transform
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, y_train)
```

```python
#Realizamos una prediccin
y_pred= algoritmo.predict(X_test)
y_pred 
```

```python
#Verifico la matriz de Confusin
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusin:')
print(matriz)
```

```python
#Calculo la precisin del modelo 
from sklearn.metrics import precision_score
precision= precision_score(y_test, y_pred, average="micro", pos_label="yes")
print('Precision del modelo:')
print(precision)
```

```python
#Calculo la exactitud del modelo
from sklearn.metrics import accuracy_score
exactitud= accuracy_score(y_test, y_pred)
print('Exactitud del modelo')
print(exactitud)
```

```python
#Calculo la sensibilidad del modelo 
from sklearn.metrics import recall_score
sensibilidad=recall_score(y_test, y_pred, average='micro', pos_label="yes")
print('Sensibilidad del modelo')
print(sensibilidad)
```
